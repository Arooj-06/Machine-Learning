# **Working with DATES**

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#now date in converted into the date and time format
```

**1. EXTRACT YEAR**

```python
date['year']=date['date'].dt.year
```

```python
date.head()
```

**2. EXTRACT MONTH**

```python
date['month']=date['date'].dt.month
#using attribute dt.month
```

```python
date.head()
```

```python
date['month_name']=date['date'].dt.month_name()
#use function dt.month_name()
```

```python
date.head()
```

**3. EXTRACT DAYS**

```python
date['day']=date['date'].dt.day
```

```python
date.head()
```

```python
date['day_name']=date['date'].dt.day_name()
```
